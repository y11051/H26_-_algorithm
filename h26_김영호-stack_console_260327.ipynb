```python
Max_size = 5
stack = []

def is_empty():
    return len(stack) == 0

def is_full():
    return len(stack) == Max_size

def push(data):
    if is_full():
        print(">stack이 차 있어서 더 이상 추가할 수 없습니다.")
    else:
        stack.append(data)

def pop():
    if is_empty():
        print(">stack이 비었습니다.")
    else:
        return stack.pop()
    return None
    return stack.pop()

def peek():
    if is_empty():
        print(">stack이 비었습니다.")
    else:
        return stack[-1]
    return None
    return stack [-1]

while True:
    print("=" * 30+"[정수형 스택 연산 실습 (용량 5)]")
    print("1. push 2. pop 3. peek 4. exit")
    print("=" * 30)

    choice = input("> 메뉴 선택: ")
    if choice == '1':
        data = int(input("> 데이터 입력: "))
        push(data)
    elif choice == '2':
        removed = pop()
        if removed is not None:
            print(f"> 가져온 데이터: {removed}")
    elif choice == '3':
        top_data = peek()
        if top_data is not None:
            print(f"> 맨 위 데이터: {top_data}")
    elif choice == '0':
        print("-" * 10 + "[정수형 스택 연산 실습 종료]"+"-"*10)
        break
    print(f"> 현재 스택 상태: {stack}\n")
    

```
